# Looping

Looping (Perulangan), merupakan sebuah logika untuk memberikan perintah perulangan. Dimana mesin akan memproses bentuk berulang, dengan batas loop yang dapat ditentukan. namun salah sintaks penulisan akan menyebabkan fungsi loop menjadi loop tidak terbatas. dalam Loop, terdapat While.... dan For.... Loop, umumnya dan if...while

```python
#   WHILE LOOP (Dasar While)

x = 1

while x <= 10:  # Selama nilai x masih kurang dari sama dengan 10
    sum = x * 2     # dimana Operasional nilai x dikali 2
    print(f"Result of multiplication of {x} is {sum}")  # mengeluarkan print dan hasil
    x += 1  # x akan bertambah 1 seiring loop berjalan hingga 10 (lihat bagian while)
    
# Variabel "sum" akan terus berulang, namun menghasilkan nilai yang berbeda, selama batas loop berakhir
```

```python
#   WHILE LOOP (Batas Variabel)

x = 10
a = 1

while a <= x:
    sum = a - 5
    print(sum)
    a += 1
```

```python
#   FOR LOOP (Dasar loop)

for i in range(1, 11, 1):   # Parameter range(Inisiasi Nilai, Limit Loop, Operasional Loop)
    print(i)
    
# Limit loop berlaku kurang dari (<)
# Jika x < 15, maka loop berhenti di angka 14
```

```python
#   FOR LOOP (Operasional Loop)

for x in range(1, 15, 1):
    sum = x * 3
    print(sum)
```

```python
# FOR LOOP (Custom Loop)

y = 10  # Inisiasi variabel dan nilai

for x in range(y):  # setiap nilai x dalam y, diberi kurung untuk menunjukkan batas loop (selalu dimulai dari 1)
    sum = x ** 2    # melakukan proses dipangkatkan 2 (**)
    print(f"Results Power of {x} is {sum}")
```

```python
#   FOR LOOP (Nested for Loop)

print("Tabel Perkalian matematika")
print()

for a in range(1, 11):  # Bagian Terluar loop (Outer Loop), untuk mengendalikan bagian Vertikal (Kolom)
    for b in range(1, 11):  # Bagian Dalam (Inner Loop), untuk mengendalikan bagian Horizontal (Baris)
        
        row_column = a * b # Mengkalikan nilai a dan b, menghasilkan perkalian di setiap baris dan kolom
        print(row_column, end="\t") # Mencetak hasil, dan memberikan fungsi tab (\t)
    
    print() # Untuk Enter, baris baru per deret perulangan
    
# Parameter range(sebagai nilai awal Akumulator(Basis nilai), Batas Deret Kolom dan Baris)
# Jika menggunakan ilmu Matriks (2D Arrays)
```

```python
#   FOR LOOP (Other Operational Case)

g = 8 # Sebagai batas Detik tempuh

print("Jarak Tempuh Kendaraan dalam 1.5 meter/sekon")
print()
for f in range(g):
    speed = f * 1.5 # Buat Operasional Sederhana, tentang waktu dan jarak tempuh

    print(f"Total jarak tempuh pada detik ke {f} adalah {speed} m")
    
# ! Perlu diingat, semua fungsi yang diletakkan di dalam Blok Loop, akan ikut Perulangan
```
