```python
#   OPERATIONAL BASIC

x = 5   # Input nilai variabel X
y = 10  # Input nilai variabel Y

sum_plus = x + y        # Operasional Penjumlahan
sum_minus = x - y       # Operasional Pengurangan
sum_divide = y / x      # Operasional Pembagian
sum_multiply = y * x    # Operasional Perkalian

print(f"Hasil Penjumlahan {x} dan {y} adalah {sum_plus}")
print(f"Hasil Pengurangan {x} dan {y} adalah {sum_minus}")
print(f"Hasil Pembagian {x} dan {y} adalah {sum_divide}")
print(f"Hasil Perkalian {x} dan {y} adalah {sum_multiply}")

#   Hasil Operasi bisa dimanipulasi lebih untuk menghasilkan Persamaan yang kompleks
```

```python
# OPERATIONAL WITH LIBRARY

import numpy as np  #   Import pustaka numpy Python, dimana dipanggil librarynya dalam nama alias "np"

a = 25
b = -20

#   Semua pemanggilan fungsi matematika tingkat lanjut seperti Kuadrat/akar, memanggil pustaka dengan aliasnya (contoh np.square)
#   Artinya kita memanggil fungsi Akar
sum_square = np.sqrt(a)     # Fungsi Akar
sum_log = np.log10(a)       # Fungsi Logaritma Basis 10
sum_abs = np.abs(b)         # Fungsi Absolut/Nilai Mutlak

print(f"Hasil Akar dari nilai {a} adalah {sum_square}")
print(f"Hasil Logaritma dari nilai {a} , dengan basis 10 adalah {sum_log}")
print(f"Hasil Absolut dari nilai {b} adalah {sum_abs}")
```

```python
# COMPLEX OPERATIONAL

f = 20
g = 10
h = 4

# Fungsi Persamaan Kompleks
sum_comp = (f * h) / g  # fungsi dalam kurung di Operasionalkan dahulu, lalu dibagi nilai g
print(sum_comp)

sum_newcomp = f * (h / g)
print(sum_newcomp)

sum_new = (f / g) ** h  # Dalam kurung di operasionalkan dulu, lalu dipangkatkan
print(sum_new)
```
